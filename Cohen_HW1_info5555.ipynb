```python
bart_data = pd.read_csv(BART_PATH)
delay_data = pd.read_csv(DELAY_PATH)
gonogo_data = pd.read_csv(GONOGO_PATH)
```

```python
def scatter(x,y):
    x_nums=[]
    for i in range(len(x)):
        x_nums.append(i)
    print (x_nums)
    x_nums = np.asarray(x_nums)
    #plt.xticks(x_nums, x)
    plt.title("performance on BART")
    plt.xlabel("time (in days)")
    plt.ylabel('average number of pumps')
    plt.axis([0,len(x_nums),1,len(y)])
    #plt.legend(loc=4)
    plt.scatter(x_nums, y)
    plt.show()
```

```python
def plot1(X):
    x_nums=[]
    y=[]
    for i in range(len(X)):
        x_nums.append(i)
        avg=0
        for j in range(1,len(X[0])-1):
            #if isinstance(X[i][j],str): 
                #if (X[i][j][0] =='[') or (X[i][j][3] =='['):
                    #X[i][j]='100'
            #print (X[i][j])
            #print (X[i][j])
            avg+=int(X[i][j])
        avg /= 4
        y.append(avg)
    #x_nums = [1,2,3]
    x_nums = np.asarray(x_nums)/10
    #plt.xticks(x_nums, x)
    plt.title("performance on delay task")
    plt.xlabel("time (in days)")
    plt.ylabel('Mean Value')
    #plt.axis([1,len(x_nums),1,len(y)])
    #plt.legend(loc=4)
    plt.plot(x_nums, y)
    plt.show()
```
